```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%pylab inline

nba = pd.read_csv("file:///C:/Users/bawinogr/Desktop/AvgHeatMap30.csv",index_col=0)




# Sort data according to Points, lowest to highest
# This was just a design choice made by Yau
# inplace=False (default) ->thanks SO user d1337
#nba_sort = nba_norm.sort('PTS', ascending=True)

#nba_sort['PTS'].head(10)
nba




```

```python
#normalize
#normalized_nba=abs((nba-nba.mean())/nba.std())
#min max normalization
normalized_nba=(nba-nba.min())/(nba.max()-nba.min())

nba=normalized_nba
# Plot it out
fig, ax = plt.subplots()
heatmap = ax.pcolor(nba, cmap=plt.cm.Reds, alpha=0.8)

# Format
fig = plt.gcf()
fig.set_size_inches(10, 5)

# turn off the frame
ax.set_frame_on(False)

# put the major ticks at the middle of each cell
ax.set_yticks(np.arange(nba.shape[0]) + 0.5, minor=False)
ax.set_xticks(np.arange(nba.shape[1]) + 0.5, minor=False)

# want a more natural, table-like display
ax.invert_yaxis()
ax.xaxis.tick_top()

labels = [
   'G','MP','PTS','TRB','AST','STL','BLK','FG%','WS','WS/48'
]

# note I could have used nba_sort.columns but made "labels" instead
ax.set_xticklabels(labels, minor=False)
ax.set_yticklabels(nba.index, minor=False)

# rotate the
plt.xticks(rotation=90)

ax.grid(False)

# Turn off all the ticks
ax = plt.gca()

for t in ax.xaxis.get_major_ticks():
    t.tick1On = False
    t.tick2On = False
for t in ax.yaxis.get_major_ticks():
    t.tick1On = False
    t.tick2On = False
    
plt.colorbar(heatmap)
```
